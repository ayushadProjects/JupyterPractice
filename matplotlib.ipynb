```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("data_plt.csv")
```

```python
plt.plot(df["Age"], df["ConvertedComp"], label = "All")
plt.xlabel("Ages")
plt.ylabel("Salary (USD)")
plt.title("Salary by Age (USD)")
plt.show()
```

```python
df_age = df.groupby("Age")
```

```python
df_group_age = pd.DataFrame(df_age["ConvertedComp"].agg(["median", "mean"]))
df_group_age.plot(title = "Median Salary by Age (USD)", ylabel = "Salary (USD)", color = ["b", "r"], grid = True)
```

```python
df_group_age = df_group_age.reset_index()
```

```python
plt.plot(df_group_age["Age"], df_group_age["mean"], color = "r", linestyle = "--", label = "Mean")
plt.plot(df_group_age["Age"], df_group_age["median"], "#5a7d9a", linewidth = 3, label = "Median")
plt.xlabel("Age")
plt.ylabel("Salary (USD)")
plt.title("Salary by Age (USD)")
plt.legend()
# plt.grid(True)
plt.style.use("seaborn-white")
# plt.xkcd()
# plt.savefig("plt.jpg")
# plt.savefig("plt.png")
plt.show()
```

```python
df_country = df[["Country", "ConvertedComp"]]
```

```python
df_group_country = df_country.groupby("Country")
df_by_country = pd.DataFrame(df_group_country["ConvertedComp"].agg(["mean", "median"]))
df_by_country.reset_index(inplace = True)
```

```python
plt.bar(df_by_country["Country"], df_by_country["mean"])
```

```python
plt.barh(df_group_age["Age"], df_group_age["median"])
plt.yticks(ticks = df_group_age["Age"], labels = df_group_age["Age"])
plt.xlabel("Salary (USD)")
plt.ylabel("Ages")
plt.title("Salary by Age (USD)")
plt.tight_layout()
plt.show()
```

```python
df_gender = pd.DataFrame(df["Gender"].value_counts().head(2))
```

```python
df_gender = df_gender.reset_index()
```

```python
df_gender.rename(columns = {"index": "Gender", "Gender": "Count"}, inplace = True)
```

```python
df_gender.sort_values(by = "Count", ascending = True, inplace = True)
```

```python
plt.barh(df_gender["Gender"], df_gender["Count"], color = ["pink", "blue"])
plt.xlabel("Count")
plt.title("Count of Gender")
```

```python
colors = ["cyan", "pink"]
explode = [0,0.2]
colors.reverse()
plt.pie( df_gender["Count"], labels = df_gender["Gender"], shadow = True, autopct = "%1.1f%%",
        explode = explode, colors = colors, wedgeprops = {"edgecolor": "black"})
plt.show()
```
