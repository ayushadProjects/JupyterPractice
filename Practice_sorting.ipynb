```python
import pandas as pd
```

```python
account = {
    "student_ID": [100, 101, 201, 331, 603],
    "first_name": ["Ayush", "Ryan", "Chisato", "Isabelle", "Ayush"],
    "last_name": ["Adhikari", "Griffin", "Nii", "Bui", "Musk"],
    "University": ["CSU", "CSU", "CSU", "DU", "Harvard"],
    "Major": ["CIS", "CS", "Piano", "Nutrition", "Economics"],
    "Salary": [200000, 100000, 50000, 75000, 250000],
    "Age": [21, 22, 30, 20, 41],
    "GPA": [4.0, 4.0, 3.5, 3.8, 4.0],
}
```

```python
df = pd.DataFrame(account)
```

```python
df.sort_values(by = "last_name")
```

```python
df.sort_values(by = "last_name", ascending = False)
```

```python
df.sort_values(by = ["GPA", "last_name"], ascending = False)
```

```python
df.sort_values(by = ["GPA", "last_name"], ascending = [False, True]) #can use inplace
```

```python
df.sort_index()
```

```python
df[["first_name", "last_name"]].sort_values(by = "first_name")
```
