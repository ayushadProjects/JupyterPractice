```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
na_vals = ["NA", "Missing", "None", "NaN"]
df = pd.read_csv("stack_overflow_data.csv", index_col = "Unnamed: 0", na_values = na_vals)
```

```python
df = df.dropna(axis = "index", how = "any", subset = ["ConvertedComp", "Age"])
```

```python
plt.plot(df["Age"], df["ConvertedComp"])
plt.show()
```

```python
filt = ( df["Age"] >= 18) & (df["Age"] <= 60) & (df["ConvertedComp"] <= 500000)
df_to_export = df.loc[filt]
```

```python
df_to_export.to_csv("data_plt.csv")
```
